```python
#2 - Analisa a quantidade de cursos oferecidos pela categoria admistrativa de cada instituição de ensino.
categoria_counts = tabela['CATEGORIA_ADMINISTRATIVA'].value_counts()

print("Contagem de Cursos por Categoria Administrativa:")
print(categoria_counts)

```

```python
#3 - Apresenta uma análise da quantidade de instituições  por região no Brasil. 


tabela_comparativa_regiao = tabela.groupby('REGIAO')['NOME_CURSO'].nunique().reset_index()

tabela_comparativa_regiao.columns = ['Região', 'Quantidade de Instituições']

print(tabela_comparativa_regiao)



import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 6))


regioes = tabela_comparativa_regiao['Região']
quantidade_cursos = tabela_comparativa_regiao['Quantidade de Instituições']

cores = ['skyblue', 'lightcoral', 'lightgreen', 'lightsalmon', 'lightseagreen']


plt.bar(regioes, quantidade_cursos, color=cores)

plt.xlabel('Região')
plt.ylabel('Quantidade de Instituições')
plt.title('Quantidade de Instituições por Região')

plt.xticks(rotation=45)


plt.tight_layout()


plt.savefig('grafico_quantidade_instituicoes_por_regiao.png', dpi=300, bbox_inches='tight')


plt.show()
```

```python
#4 - Verificando quantidade de cursos oferecidos por Instituição e total de Instituições.

contagem_instituicoes = tabela['NOME_IES'].value_counts().reset_index()

contagem_instituicoes.columns = ['Nome da Instituição', 'Quantidade de Cursos']

total_instituicoes = tabela['NOME_IES'].nunique()

print("Total de Instituições:", total_instituicoes)

display(contagem_instituicoes)




```

```python
#7-Analisando a disponibilidade do curso por região, considerando a quantidade de instituição por região e em quais cidades este curso está disponível.
curso_procurado = "MEDICINA"

curso_filtrado = tabela[tabela['NOME_CURSO'] == curso_procurado]

# Analisar a quantidade de cidades que oferecem o curso
cidades_do_curso = curso_filtrado['MUNICIPIO'].unique()
quantidade_de_cidades = len(cidades_do_curso)
print("Quantidade de cidades onde o curso está disponível:", quantidade_de_cidades)


# Analisar a quantidade de cidades por região que oferecem o curso
cidades_por_regiao = curso_filtrado.groupby('REGIAO')['MUNICIPIO'].nunique()

print("Quantidade de cidades por região que oferecem o curso:")
print(cidades_por_regiao)

# Analisar a quantidade de instituições de ensino por região que oferecem o curso
instituicoes_por_regiao = curso_filtrado.groupby(['REGIAO', 'NOME_IES'])['NOME_IES'].nunique().groupby('REGIAO').count()
print("Quantidade de instituições que oferecem o curso por região:")
print(instituicoes_por_regiao)


plt.figure(figsize=(3, 3))
plt.pie(instituicoes_por_regiao, labels=instituicoes_por_regiao.index, autopct='%1.1f%%', startangle=140)
plt.title(f"Distribuição de Instituições de Ensino por Região para o Curso '{curso_procurado}'")
plt.axis('equal') 


plt.savefig('grafico_qt_instituicao_por_regiaoMED.png', dpi=300, bbox_inches='tight')

plt.show()



```

```python

planilha2020 = pd.read_excel('PERC_DOCENTES_REGIAO_UF_2020.xlsx', header=[7, 8, 9], skiprows=[10])
planilha2020.columns = planilha2020.columns.droplevel([0, 1])



planilha2021 = pd.read_excel('PERC_DOCENTES_REGIAO_UF_2021.xlsx', header=[7, 8, 9], skiprows=[10])
planilha2021.columns = planilha2021.columns.droplevel([0, 1])

def salvar_grafico_como_png(plt, nome_arquivo):
    plt.savefig(nome_arquivo, bbox_inches='tight', dpi=300)
    plt.close() 

cores_mestrado = ['blue', 'green', 'red', 'purple', 'orange']
cores_doutorado = ['lightblue', 'lightgreen', 'salmon', 'violet', 'lightcoral']

grupo_regiao_2020 = planilha2020.groupby('NO_REGIAO')
grupo_regiao_2021 = planilha2021.groupby('NO_REGIAO')

media_percentual_mestrado_2020 = grupo_regiao_2020['PERCENTUAL_MESTRADO'].mean()
media_percentual_doutorado_2020 = grupo_regiao_2020['PERCENTUAL_DOUTORADO'].mean()

media_percentual_mestrado_2021 = grupo_regiao_2021['PERCENTUAL_MESTRADO'].mean()
media_percentual_doutorado_2021 = grupo_regiao_2021['PERCENTUAL_DOUTORADO'].mean()

medias_por_regiao_2020 = pd.DataFrame({
    'Região': media_percentual_mestrado_2020.index,
    'Média Percentual Mestrado 2020': media_percentual_mestrado_2020.values,
    'Média Percentual Doutorado 2020': media_percentual_doutorado_2020.values
})

medias_por_regiao_2021 = pd.DataFrame({
    'Região': media_percentual_mestrado_2021.index,
    'Média Percentual Mestrado 2021': media_percentual_mestrado_2021.values,
    'Média Percentual Doutorado 2021': media_percentual_doutorado_2021.values
})


medias_comparacao = medias_por_regiao_2020.merge(medias_por_regiao_2021, on='Região')

plt.figure(figsize=(8, 4))

bar_width = 0.35
index = range(len(medias_comparacao))

plt.bar(index, medias_comparacao['Média Percentual Mestrado 2020'], bar_width, label='Mestrado 2020', color='blue', alpha=0.7)
plt.bar([i + bar_width for i in index], medias_comparacao['Média Percentual Mestrado 2021'], bar_width, label='Mestrado 2021', color='green', alpha=0.7)

plt.xlabel('Região')
plt.ylabel('Média Percentual de Mestrado')
plt.title('Comparação da Média Percentual de Mestrado por Região em 2020 e 2021')
plt.xticks([i + bar_width / 2 for i in index], medias_comparacao['Região'], rotation=45)
plt.legend()
plt.savefig('grafico_mestrado20202021.png', dpi=300, bbox_inches='tight')
plt.tight_layout()
plt.show()







plt.figure(figsize=(8, 4))

bar_width = 0.35
index = range(len(medias_comparacao))


cor_doutorado_2020 = 'blue'
cor_doutorado_2021 = 'green'

plt.bar(index, medias_comparacao['Média Percentual Doutorado 2020'], bar_width, label='Doutorado 2020', color=cor_doutorado_2020, alpha=0.7)
plt.bar([i + bar_width for i in index], medias_comparacao['Média Percentual Doutorado 2021'], bar_width, label='Doutorado 2021', color=cor_doutorado_2021, alpha=0.7)

plt.xlabel('Região')
plt.ylabel('Média Percentual de Doutorado')
plt.title('Comparação da Média Percentual de Doutorado por Região em 2020 e 2021')
plt.xticks([i + bar_width / 2 for i in index], medias_comparacao['Região'], rotation=45)
plt.legend()

plt.savefig('grafico_doutorado20202021.png', dpi=300, bbox_inches='tight')

plt.tight_layout()

plt.show()









```
